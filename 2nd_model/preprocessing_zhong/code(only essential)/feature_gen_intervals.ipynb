```python
import pandas as pd
```

```python
orders_bef_dec = pd.read_csv('orders_before_jan_new.csv', delimiter=',')
```

```python
orders_bef_dec
```

```python
orders_bef_dec['comID'] = orders_bef_dec['userID'].astype(str) + orders_bef_dec['itemID'].astype(str)
```

```python
orders_bef_dec
```

```python
groups = orders_bef_dec.groupby('comID')
```

```python
se = groups['intervals'].sum() + 1
```

```python
orders_bef_dec = pd.merge(orders_bef_dec, se, on='comID', how='left')
```

```python
orders_bef_dec.drop(['intervals_x'], axis=1, inplace=True)
```

```python
orders_bef_dec
```

```python
orders_bef_dec_label = pd.read_csv('orders_before_jan_labeled.csv', delimiter=',')
```

```python
orders_bef_dec['label'] = orders_bef_dec_label['label']
```

```python
orders_bef_dec.drop('comID', axis=1, inplace=True)
```

```python
orders_bef_dec
```

```python
orders_bef_dec.to_csv('orders_bef_jan_intervalsY_labeled.csv', sep=',', encoding='utf-8', index=False)
```
