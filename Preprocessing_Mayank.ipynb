```python
# reading the items file
df_items = pd.read_csv('C:/Users/mayan/Downloads/SpringSemester 2022/Data Mining 2/Data Mining Project/DMC_2022_task/items.csv',sep='|',error_bad_lines=False)

print("Shape of df: ",df_items.shape)
print("Info of df: ",df_items.info())
print("Describe df: ",df_items.describe())
```

```python
# selecting the itemID and categories column 

item_categories= df_items[["itemID","brand","categories"]]

# flattening the categories by item IDs
item_categories['categories']=item_categories['categories'].str.strip('[]').str.split(',') 
item_categories = item_categories.explode('categories')

display(item_categories)
```

```python
# renaming the column categories to column
item_categories.rename(columns={'categories': 'category'}, inplace=True)

# fill NA values in category column -- 4300 is the category with null values
item_categories["category"] = item_categories["category"].fillna(4300)

# converting the datatype of category column to int64
item_categories["category"] = item_categories["category"].astype('float').astype('int64')

# shape of itemcategories
print("Describe df: ",item_categories.info())
display(item_categories)
```

```python
# Joining the dataframe to category_hierarchy file
#df_item_cat = pd.concat([item_categories,df_cat_hierarchy],axis=1,keys=['itemID','category']).reset_index()
df_item_cat = pd.merge(item_categories, df_cat_hierarchy,how="inner",on=["category"])

display(df_item_cat)

# shape of dataframe
print("Describe df: ",df_item_cat.info())
```
