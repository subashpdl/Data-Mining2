```python
import pandas as pd
```

```python
orders = pd.read_csv('orders.csv', delimiter='|')
```

```python
#orders.dtypes
```

```python
orders['date'] = pd.to_datetime(orders['date'])
```

```python
#orders.dtypes
```

```python
jun_s = pd.to_datetime('2020-06-01')
jun_e = pd.to_datetime('2020-06-30')

jul_s = pd.to_datetime('2020-07-01')
jul_e = pd.to_datetime('2020-07-31')

aug_s = pd.to_datetime('2020-08-01')
aug_e = pd.to_datetime('2020-08-31')

sep_s = pd.to_datetime('2020-09-01')
sep_e = pd.to_datetime('2020-09-30')

oct_s = pd.to_datetime('2020-10-01')
oct_e = pd.to_datetime('2020-10-31')

nov_s = pd.to_datetime('2020-11-01')
nov_e = pd.to_datetime('2020-11-30')

dec_s = pd.to_datetime('2020-12-01')
dec_e = pd.to_datetime('2020-12-31')

jan_s = pd.to_datetime('2021-01-01')
jan_e = pd.to_datetime('2021-01-31')
```

```python
orders_jun = orders[(orders['date'] >= jun_s) & (jun_e >= orders['date'])]

orders_jul = orders[(orders['date'] >= jul_s) & (jul_e >= orders['date'])]

orders_aug = orders[(orders['date'] >= aug_s) & (aug_e >= orders['date'])]

orders_sep = orders[(orders['date'] >= sep_s) & (sep_e >= orders['date'])]

orders_oct = orders[(orders['date'] >= oct_s) & (oct_e >= orders['date'])]

orders_nov = orders[(orders['date'] >= nov_s) & (nov_e >= orders['date'])]

orders_dec = orders[(orders['date'] >= dec_s) & (dec_e >= orders['date'])]

orders_jan = orders[(orders['date'] >= jan_s) & (jan_e >= orders['date'])]
```

```python
orders_jun
```

```python
dic = {'jun':30, 'jul':31, 'aug':31, 'sep':30, 'oct':31, 'nov':30, 'dec':31, 'jan':31}
```

```python
groups_jun = list(orders_jun.groupby('itemID'))

ls = list()

for group in groups_jun:
    ls.append((group[0], group[1].shape[0]))
    
#create df with itemID, purchassedTime at june, purchased frequency in june
df_item_purchased_jun = pd.DataFrame(ls, columns=["itemID", "purchased_times"])
df_item_purchased_jun['purchased_frequency'] = df_item_purchased_jun['purchased_times'] / dic['jun']
```

```python
df_item_purchased_jun
```

```python
orders_jun = pd.merge(orders_jun, df_item_purchased_jun, on='itemID', how='left')
```

```python
orders_jun.drop('purchased_times', axis=1, inplace=True)
```
