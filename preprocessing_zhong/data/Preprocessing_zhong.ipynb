# #orders tables feature generating

```python
#below six new features are generated, namely
#user_total_purchase_times, user_total_orders, item_total_purchase_time, item_total_orders, weekday and date of the month
#user_total_purchase_times, user_total_orders, item_total_purchase_time, item_total_orders are scaled to one month average
#for orders_bef_dec/jan since there are 6/7 months for them respectively
```

```python
orders_bef_dec = pd.read_csv('orders_bef_dec.csv',',')
orders_bef_jan = pd.read_csv('orders_bef_jan.csv',',')
orders_dec = pd.read_csv('orders_dec.csv',',')
orders_jan = pd.read_csv('orders_jan.csv',',')
```

```python
#orders_jan
```

```python
#group by user
groups1 = orders_bef_dec.groupby('userID')
groups2 = orders_bef_jan.groupby('userID')
groups3 = orders_dec.groupby('userID')
groups4 = orders_jan.groupby('userID')
```

```python
groups1 = list(groups1)
groups2 = list(groups2)
groups3 = list(groups3)
groups4 = list(groups4)
```

```python
ls1 = list()
ls2 = list()
ls3 = list()
ls4 = list()
```

```python
#list [('userID', 'user_total_purchase_time', 'user_total_orders'),...]
#total_purchase_time, total_orders are not scaled
for group in groups1:
    ls1.append((group[0], group[1].shape[0], group[1]['order'].sum()))
```

```python
for group in groups2:
    ls2.append((group[0], group[1].shape[0], group[1]['order'].sum()))
```

```python
for group in groups3:
    ls3.append((group[0], group[1].shape[0], group[1]['order'].sum()))
```

```python
for group in groups4:
    ls4.append((group[0], group[1].shape[0], group[1]['order'].sum()))
```

```python
df1 = pd.DataFrame(ls1, columns=["userID", "user_total_purchase_time","user_total_orders"])
df2 = pd.DataFrame(ls2, columns=["userID", "user_total_purchase_time","user_total_orders"])
df3 = pd.DataFrame(ls3, columns=["userID", "user_total_purchase_time","user_total_orders"])
df4 = pd.DataFrame(ls4, columns=["userID", "user_total_purchase_time","user_total_orders"])
```

```python
#append user_total_purchase_time, user_total_orders to orders...
orders_bef_dec = pd.merge(orders_bef_dec, df1, on='userID', how='left')
orders_bef_jan = pd.merge(orders_bef_jan, df2, on='userID', how='left')
orders_dec = pd.merge(orders_dec, df3, on='userID', how='left')
orders_jan = pd.merge(orders_jan, df4, on='userID', how='left')
```

```python
#group by item
groups1 = orders_bef_dec.groupby('itemID')
groups2 = orders_bef_jan.groupby('itemID')
groups3 = orders_dec.groupby('itemID')
groups4 = orders_jan.groupby('itemID')
```

```python
groups1 = list(groups1)
groups2 = list(groups2)
groups3 = list(groups3)
groups4 = list(groups4)
```

```python
ls1 = list()
ls2 = list()
ls3 = list()
ls4 = list()
```

```python
#list [('itemID', 'item_total_purchase_time', 'item_total_orders'),...]
#total_purchase_time, total_orders are not scaled
for group in groups1:
    ls1.append((group[0], group[1].shape[0], group[1]['order'].sum()))
```

```python
for group in groups2:
    ls2.append((group[0], group[1].shape[0], group[1]['order'].sum()))
```

```python
for group in groups3:
    ls3.append((group[0], group[1].shape[0], group[1]['order'].sum()))
```

```python
for group in groups4:
    ls4.append((group[0], group[1].shape[0], group[1]['order'].sum()))
```

```python
df1 = pd.DataFrame(ls1, columns=["itemID", "item_total_purchase_time","item_total_orders"])
df2 = pd.DataFrame(ls2, columns=["itemID", "item_total_purchase_time","item_total_orders"])
df3 = pd.DataFrame(ls3, columns=["itemID", "item_total_purchase_time","item_total_orders"])
df4 = pd.DataFrame(ls4, columns=["itemID", "item_total_purchase_time","item_total_orders"])
```

```python
#append item_total_purchase_time, item_total_orders into orders...
orders_bef_dec = pd.merge(orders_bef_dec, df1, on='itemID', how='left')
orders_bef_jan = pd.merge(orders_bef_jan, df2, on='itemID', how='left')
orders_dec = pd.merge(orders_dec, df3, on='itemID', how='left')
orders_jan = pd.merge(orders_jan, df4, on='itemID', how='left')
```

```python
#scale the user_total_purchase_time
orders_bef_dec['user_total_purchase_time'] = (orders_bef_dec['user_total_purchase_time'] / 6).round(2)
orders_bef_dec['user_total_orders'] = (orders_bef_dec['user_total_orders'] / 6).round(2)
orders_bef_dec['item_total_purchase_time'] = (orders_bef_dec['item_total_purchase_time'] / 6).round(2)
orders_bef_dec['item_total_orders'] = (orders_bef_dec['item_total_orders'] / 6).round(2)
```

```python
#scale the user_total_purchase_time
orders_bef_jan['user_total_purchase_time'] = (orders_bef_jan['user_total_purchase_time'] / 7).round(2)
orders_bef_jan['user_total_orders'] = (orders_bef_jan['user_total_orders'] / 7).round(2)
orders_bef_jan['item_total_purchase_time'] = (orders_bef_jan['item_total_purchase_time'] / 7).round(2)
orders_bef_jan['item_total_orders'] = (orders_bef_jan['item_total_orders'] / 7).round(2)
```

```python
orders_bef_dec
```

```python
orders_bef_dec['date'] = pd.to_datetime(orders_bef_dec['date'])
orders_bef_jan['date'] = pd.to_datetime(orders_bef_jan['date'])
orders_dec['date'] = pd.to_datetime(orders_dec['date'])
orders_jan['date'] = pd.to_datetime(orders_jan['date'])
```

```python
#generate day of the week
ls1_weekday = list()
ls2_weekday = list()
ls3_weekday = list()
ls4_weekday = list()

for i in range(orders_bef_dec.shape[0]):
    ls1_weekday.append(orders_bef_dec.iloc[i]['date'].weekday())
    
for i in range(orders_bef_jan.shape[0]):
    ls2_weekday.append(orders_bef_jan.iloc[i]['date'].weekday())
    
for i in range(orders_dec.shape[0]):
    ls3_weekday.append(orders_dec.iloc[i]['date'].weekday())
    
for i in range(orders_jan.shape[0]):
    ls4_weekday.append(orders_jan.iloc[i]['date'].weekday())
```

```python
#append day of the week
orders_bef_dec['weekday'] = ls1_weekday
orders_bef_jan['weekday'] = ls2_weekday
orders_dec['weekday'] = ls3_weekday
orders_jan['weekday'] = ls4_weekday
```

```python
#generate date of the month
ls1_date_of_month = list()
ls2_date_of_month = list()
ls3_date_of_month = list()
ls4_date_of_month = list()

for i in range(orders_bef_dec.shape[0]):
    ls1_date_of_month.append(orders_bef_dec.iloc[i]['date'].day)
    
for i in range(orders_bef_jan.shape[0]):
    ls2_date_of_month.append(orders_bef_jan.iloc[i]['date'].day)
    
for i in range(orders_dec.shape[0]):
    ls3_date_of_month.append(orders_dec.iloc[i]['date'].day)
    
for i in range(orders_jan.shape[0]):
    ls4_date_of_month.append(orders_jan.iloc[i]['date'].day)
```

```python
#append date of the month
orders_bef_dec['date_of_month'] = ls1_date_of_month
orders_bef_jan['date_of_month'] = ls2_date_of_month
orders_dec['date_of_month'] = ls3_date_of_month
orders_jan['date_of_month'] = ls4_date_of_month
```

```python
orders_bef_dec.to_csv('orders_bef_dec_fea_gen.csv',index=False,encoding='utf-8',sep=',')
orders_bef_jan.to_csv('orders_bef_jan_fea_gen.csv',index=False,encoding='utf-8',sep=',')
orders_dec.to_csv('orders_dec_fea_gen.csv',index=False,encoding='utf-8',sep=',')
orders_jan.to_csv('orders_jan_fea_gen.csv',index=False,encoding='utf-8',sep=',')
```
