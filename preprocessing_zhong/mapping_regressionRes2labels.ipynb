```python
import pandas as pd
```

```python
#pls change the file below to the output file of regression models
#the following two files opened are just for helping with creating this mapping file
prediction = pd.read_csv('orders_dec_fea_gen.csv', ',')
pred = pd.read_csv('orders_jan_fea_gen.csv', ',')
```

```python
#prediction.dtypes
```

```python
#transform date from object to datetime
prediction['date'] = pd.to_datetime(prediction['date'])
pred['date'] = pd.to_datetime(pred['date'])
```

```python
#prediction.dtypes
```

```python
#get date as a list
ls_dec_date = prediction['date']
ls_jan_date = pred['date']
#get label as a list
ls_dec_days = prediction['label']
ls_jan_days = pred['label']
```

```python
#create a list to store predicted/expected next purchase/replenishment date
ls_dec_pre_next_purchase_date = list()
ls_jan_pre_next_purchase_date = list()
```

```python
for i in range(len(ls_dec_date)):
    ls_dec_pre_next_purchase_date.append(ls_dec_date[i] + pd.Timedelta(days=ls_dec_days[i]))
    
for i in range(len(ls_jan_date)):
    ls_jan_pre_next_purchase_date.append(ls_jan_date[i] + pd.Timedelta(days=ls_jan_days[i]))
```

```python
#ls_dec_pre_next_purchase_date
```

```python
#ls_dec_date
```

```python
#ls_dec_days
```

```python
#transform predicted next purachse/replenishment date into weeks(1st week, 2rd week...)
label_dec_in_week = list()
label_jan_in_week = list()
```

```python
dew_s = pd.to_datetime('2020-12-01')
dew_e = pd.to_datetime('2020-12-07')

dzw_s = pd.to_datetime('2020-12-08')
dzw_e = pd.to_datetime('2020-12-14')

ddw_s = pd.to_datetime('2020-12-15')
ddw_e = pd.to_datetime('2020-12-21')

dvw_s = pd.to_datetime('2020-12-22')
dvw_e = pd.to_datetime('2020-12-28')

jew_s = pd.to_datetime('2021-01-01')
jew_e = pd.to_datetime('2021-01-07')

jzw_s = pd.to_datetime('2021-01-08')
jzw_e = pd.to_datetime('2021-01-14')

jdw_s = pd.to_datetime('2021-01-15')
jdw_e = pd.to_datetime('2021-01-21')

jvw_s = pd.to_datetime('2021-01-22')
jvw_e = pd.to_datetime('2021-01-28')
```

```python
for i in range(len(ls_dec_pre_next_purchase_date)):
    if((ls_dec_pre_next_purchase_date[i]>=dew_s) & (ls_dec_pre_next_purchase_date[i]<=dew_e)):
        label_dec_in_week.append(1)
    elif((ls_dec_pre_next_purchase_date[i]>=dzw_s) & (ls_dec_pre_next_purchase_date[i]<=dzw_e)):
        label_dec_in_week.append(2)
    elif((ls_dec_pre_next_purchase_date[i]>=ddw_s) & (ls_dec_pre_next_purchase_date[i]<=ddw_e)):
        label_dec_in_week.append(3)
    elif((ls_dec_pre_next_purchase_date[i]>=dvw_s) & (ls_dec_pre_next_purchase_date[i]<=dvw_e)):
        label_dec_in_week.append(4)
    else:
        label_dec_in_week.append(0)
        
for i in range(len(ls_jan_pre_next_purchase_date)):
    if((ls_jan_pre_next_purchase_date[i]>=jew_s) & (ls_jan_pre_next_purchase_date[i]<=jew_e)):
        label_jan_in_week.append(1)
    elif((ls_jan_pre_next_purchase_date[i]>=jzw_s) & (ls_jan_pre_next_purchase_date[i]<=jzw_e)):
        label_jan_in_week.append(2)
    elif((ls_jan_pre_next_purchase_date[i]>=jdw_s) & (ls_jan_pre_next_purchase_date[i]<=jdw_e)):
        label_jan_in_week.append(3)
    elif((ls_jan_pre_next_purchase_date[i]>=jvw_s) & (ls_jan_pre_next_purchase_date[i]<=jvw_e)):
        label_jan_in_week.append(4)
    else:
        label_jan_in_week.append(0)
```

```python
prediction['label'] = label_dec_in_week
pred['label'] = label_jan_in_week
```
