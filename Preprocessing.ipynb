```python
#Load the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

# Data Preprocessing

```python
# reading the category hierarchy file
df_cat_hierarchy = pd.read_csv('category_hierarchy.csv', delimiter='|',error_bad_lines=False)

print("Shape of df: ",df_cat_hierarchy.shape)
print("Info of df: ",df_cat_hierarchy.info())
print("Describe df: ",df_cat_hierarchy.describe())
```

```python
# reading the items file
df_items = pd.read_csv('items.csv',sep='|',error_bad_lines=False)

print("Shape of df: ",df_items.shape)
print("Info of df: ",df_items.info())
print("Describe df: ",df_items.describe())
```

```python
#reading order file
df_orders = pd.read_csv('orders.csv', delimiter='|')

print("Shape of df: ",df_items.shape)
print("Info of df: ",df_items.info())
print("Describe df: ",df_items.describe())
```

```python
# flattening the categories by item IDs
df_items['categories'] = df_items['categories'].str.strip('[]').str.split(',') 
df_items = df_items.explode('categories')

df_items.head(5)
```

```python
# renaming the column categories to column
df_items.rename(columns={'categories': 'category'}, inplace=True)

# fill NA values in category column -- 4300 is the category with null values
df_items["category"] = df_items["category"].fillna(4300)

# converting the datatype of category column to int64
df_items["category"] = df_items["category"].astype('float').astype('int64')

# shape of itemcategories
print("Describe df: ",df_items.info())
display(df_items)
```

```python
# Joining the dataframe to category_hierarchy file
#df_item_cat = pd.concat([item_categories,df_cat_hierarchy],axis=1,keys=['itemID','category']).reset_index()
df_item_cat = pd.merge(df_items, df_cat_hierarchy,how="inner",on=["category"])
df_item_cat

display(df_item_cat)

# shape of dataframe
print("Describe df: ",df_item_cat.info())
```

```python
# Joining df_orders dataframe and df_items dataframe
df_item_order = pd.merge(df_item_cat, df_orders,how="inner",on=["itemID"])
display(df_item_order)
#change the date coulumn's datayte to datetime datatype
df_item_order['date'] = pd.to_datetime(df_item_order['date'])
# shape of dataframe
print("Describe df: ",df_item_order.info())
```
