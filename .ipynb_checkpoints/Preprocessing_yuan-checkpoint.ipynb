```python
#Load the libraries
import numpy as np
import pandas as pd
```

# Data Preprocessing

```python
# reading the category hierarchy file
df_cat_hierarchy = pd.read_csv('category_hierarchy.csv', delimiter='|',error_bad_lines=False)

print("Shape of df: ",df_cat_hierarchy.shape)
print("***********************************************")
print("Info of df: ",df_cat_hierarchy.info())
print("***********************************************")
print("Describe df: ",df_cat_hierarchy.describe())
```

```python
# reading the items file
df_items = pd.read_csv('items.csv',sep='|',error_bad_lines=False)

print("Shape of df: ",df_items.shape)
print("***********************************************")
print("Info of df: ",df_items.info())
print("***********************************************")
print("Describe df: ",df_items.describe())
```

```python
#reading orders_before_dec_new file with additional attribute 'intervals'
#'intervals' indicates how many days after the first purchase the user made a second purchase with regarding to the same item

df_orders_dec = pd.read_csv('orders_before_dec_new.csv', delimiter=',')

print("Shape of df: ",df_orders_dec.shape)
print("***********************************************")
print("Info of df: ",df_orders_dec.info())
print("***********************************************")
print("Describe df: ",df_orders_dec.describe())
```

```python
#reading orders_before_jan_new file with additional attribute 'intervals'
#'intervals' indicates how many days after the first purchase the user made a second purchase with regarding to the same item
#-1 indicates no re-purchase

df_orders_jan = pd.read_csv('orders_before_jan_new.csv', delimiter=',')

print("Shape of df: ",df_orders_jan.shape)
print("***********************************************")
print("Info of df: ",df_orders_jan.info())
print("***********************************************")
print("Describe df: ",df_orders_jan.describe())
```

```python
# flattening the categories by item IDs
df_items['categories'] = df_items['categories'].str.strip('[]').str.split(',') 
df_items = df_items.explode('categories')

df_items.head(5)
```

```python
# renaming the column categories to column
df_items.rename(columns={'categories': 'category'}, inplace=True)

# fill NA values in category column -- 4300 is the category with null values
df_items["category"] = df_items["category"].fillna(4300)

# converting the datatype of category column to int64
df_items["category"] = df_items["category"].astype('float').astype('int64')

# shape of itemcategories
print("Describe df: ",df_items.info())
display(df_items)
```

```python
# Joining the dataframe to category_hierarchy file
df_item_cat = pd.merge(df_items, df_cat_hierarchy,how="inner",on=["category"])
df_item_cat

display(df_item_cat)

# shape of dataframe
print("Describe df: ",df_item_cat.info())
```

```python
# Joining df_orders_dec dataframe and df_items dataframe
df_item_order_dec = pd.merge(df_item_cat, df_orders_dec,how="inner",on=["itemID"])
display(df_item_order_dec)
#change the date coulumn's datayte to datetime datatype
df_item_order_dec['date'] = pd.to_datetime(df_item_order_dec['date'])
# shape of dataframe
print("Describe df: ",df_item_order_dec.info())
```

```python
# Joining df_orders_jan dataframe and df_items dataframe
df_item_order_jan = pd.merge(df_item_cat, df_orders_jan,how="inner",on=["itemID"])
display(df_item_order_jan)
#change the date coulumn's datayte to datetime datatype
df_item_order_jan['date'] = pd.to_datetime(df_item_order_jan['date'])
# shape of dataframe
print("Describe df: ",df_item_order_jan.info())
```
